```python
import seaborn as sns
```

```python
a_df = sns.countplot(x='Size' ,data=data)
for bar in a_df.containers:
    a_df.bar_label(bar)
```

```python
data.columns
```

```python
data.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty', ascending=0)
```

```python
data.groupby(['Category'] , as_index=False)['Qty'].sum().sort_values(by='Qty' , ascending=0)
```

```python
df=sns.countplot(data=data,x='Category')
for bar in df.containers:
    df.bar_label(bar)

plt.xticks(rotation=90)

plt.show()
```

```python
data.groupby(['ship-state'], as_index=False)['Qty'].sum().sort_values(by = 'Qty',ascending=False)
```

```python
data['ship-state'].value_counts()
```

```python
plt.figure(figsize=(15,6))
sns.countplot(data=data, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('distribution of state')
plt.xticks(rotation=90)
plt.show()

```

```python
df=data[data['ship-state']=='MAHARASHTRA'].groupby('ship-city')['Qty'].sum().sort_values(ascending=False).reset_index().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x='ship-city',y='Qty', data=df)
plt.xlabel('City')
plt.ylabel('Total Quantity')
plt.title('Top 10 Cities in Maharashtra by Total Quantity Ordered')
plt.xticks(rotation=80, ha='right')

plt.show()
```
